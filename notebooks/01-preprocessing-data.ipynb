```python
DATA_DIR = '../data/'
```

```python
import os

import pandas as pd
import numpy as np
```

```python
os.path.join(DATA_DIR, 'marketing_campaign.xls')
```

```python
df = pd.read_csv(os.path.join(DATA_DIR, 'marketing_campaign.xls'), sep='\t')
df.head()
```

```python
df.isnull().sum()
```

```python
df.loc[df.isnull().sum(axis=1) > 0, 'Income'] = 0
```

```python
df.describe()
```

```python
df.select_dtypes(include='object')
```

```python
df_new = df.copy(deep=True)
df_new = pd.concat([df_new, pd.get_dummies(df.Education)], axis=1)
df_new = pd.concat([df_new, pd.get_dummies(df.Marital_Status)], axis=1)
df_new = df_new.drop(columns=['Education', 'Marital_Status'])
df_new.head()
```

```python
df_new.Dt_Customer = pd.to_datetime(df.Dt_Customer, dayfirst=True)
df_new.head()
```

```python
df_new.info()
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(6,7, figsize=(20,16))
mean_resample = df_new.resample(on='Dt_Customer', rule='1Q').mean()
for col, ax in zip(mean_resample, axes.ravel()):
    mean_resample[col].plot(ax=ax)
plt.show()
```

```python
df_new.Dt_Customer = pd.to_numeric(df_new.Dt_Customer, downcast="unsigned")
df_new.Dt_Customer = df_new.Dt_Customer - df_new.Dt_Customer.min()
df_new.Dt_Customer
```

```python
df_new = df_new.drop_duplicates()
```

```python
df_new = df_new.drop(columns=['ID'])
df_new.to_csv(os.path.join(DATA_DIR, 'preprocessed.csv'), index=False)
```
